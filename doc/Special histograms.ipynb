```python
# Necessary import evil
%matplotlib inline
from physt import histogram, binning, special
import numpy as np
import matplotlib.pyplot as plt
```

## Polar histogram

```python
# Generate some points in the Cartesian coordinates
x = np.random.rand(10000)  
y = np.random.rand(10000)
```

```python
hist = special.polar_histogram(x, y, 30, 50)
hist
```

```python
# Standard axes names
hist.axis_names
```

```python
# Simple plotting, similar to Histogram2D
hist.plot(density=True, show_zero=False, cmap="rainbow", lw=0.5, figsize=(10, 10));
```

```python
hist.projection("r").plot(cmap="rainbow")
```

```python
hist.projection("phi").plot(cmap="rainbow")
```
