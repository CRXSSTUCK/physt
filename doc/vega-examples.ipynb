```python
from physt.examples import normal_h1, iris_h1, normal_h2, normal_h3, munros
from physt.plotting import set_default_backend

set_default_backend("vega")
```

```python
H = normal_h1()
H.plot.scatter()
```

```python
H.plot.bar(cumulative=True, xlabel="Other label")
```
